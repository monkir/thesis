```python
import pandas as pd
import numpy as np
import os
import pickle
```

```python
def pickle_load(dir, filename):
  # open a file, where you stored the pickled data
  file = open(os.path.join(dir,filename), 'rb')

  # dump information to that file
  data = pickle.load(file)

  # close the file
  file.close()
  return data
```

```python
data = pd.read_csv("predictions 3.2.csv")
label = pickle_load("", "test_data_y.pk")
```

```python
data = np.array(data)
label = np.array(label)
print(data.shape, label.shape)
```

```python
# Deleting first colum (Data serial)
data = data[:, 1:]
print(data.shape, label.shape)
```

```python
# applying Information Gain
from skfeature.function.similarity_based import fisher_score
importance = fisher_score.fisher_score(X=data, y=label)
print(importance.shape)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(15,5))
plt.plot()
plt.style.use('_mpl-gallery')

# make data:
x = 0.5 + np.arange(len(importance))

# plot
# fig, ax = plt.subplots()

plt.bar(x, importance[:], width=1, edgecolor="white", linewidth=0.7)

# ax.set(xlim=(0, 8), xticks=np.arange(1, 10),
#        ylim=(0, 1), yticks=np.arange(1, 10))

plt.show()
```
