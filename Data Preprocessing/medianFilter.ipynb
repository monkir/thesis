```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
import tqdm
import math
```

```python
from PIL import Image, ImageFilter
```

```python
dir = 'Avatar'
imgData = []
for file in os.listdir(dir):
    filepath = os.path.join(dir, file)
    img = cv2.imread(filepath)
    try:
        img = cv2.resize(img, (300,300))
        # print(len(img))
        imgData.append(img)
    except:
        print("An exception occurred")
imgData = np.array(imgData)
print(imgData.shape)
```

```python
plt.figure(figsize=(15,8))
col = 5
row = math.ceil(len(imgData)/col)
for i, img in enumerate(imgData):
    plt.subplot(row, col, i+1)
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_RGB2BGR))
plt.show()
```

```python
k5_imgData = []
k9_imgData = []
for img in imgData:
    k5_imgData.append(cv2.medianBlur(img, ksize=5))
    k9_imgData.append(cv2.medianBlur(img, ksize=9))
k5_imgData = np.array(k5_imgData)
k9_imgData = np.array(k9_imgData)
print(k5_imgData.shape)
print(k9_imgData.shape)
```

```python
plt.figure(figsize=(15,8))
for i in range(5):
    plt.subplot(3, 5, i+1)
    plt.imshow(cv2.cvtColor(imgData[i], cv2.COLOR_RGB2BGR))
    plt.subplot(3, 5, i+6)
    plt.imshow(cv2.cvtColor(k5_imgData[i], cv2.COLOR_RGB2BGR))
    plt.subplot(3, 5, i+11)
    plt.imshow(cv2.cvtColor(k9_imgData[i], cv2.COLOR_RGB2BGR))
plt.show()
```
