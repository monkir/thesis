```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
import tqdm
import math
```

```python
dir = 'Avatar'
imgData = []
for file in os.listdir(dir):
    filepath = os.path.join(dir, file)
    img = cv2.imread(filepath)
    try:
        img = cv2.resize(img, (300,300))
        # print(len(img))
        imgData.append(img)
    except:
        print("An exception occurred")
imgData = np.array(imgData)
print(imgData.shape)
```

```python
plt.figure(figsize=(15,8))
col = 5
row = math.ceil(len(imgData)/col)
for i, img in enumerate(imgData):
    plt.subplot(row, col, i+1)
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_RGB2BGR))
plt.show()
```

```python
avg_img = np.mean(imgData, axis=0)
avg_img = avg_img.astype('uint8')
print(avg_img.shape)
```

```python
plt.imshow(avg_img)
```

```python
norm_imgData = imgData - avg_img
print(norm_imgData.shape)
```

```python
plt.figure(figsize=(15,8))
col = 5
row = math.ceil(len(imgData)/col)
for i, img in enumerate(norm_imgData):
    plt.subplot(row, col, i+1)
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_RGB2BGR))
plt.show()
```
